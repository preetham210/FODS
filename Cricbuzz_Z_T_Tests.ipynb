```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
cric=pd.read_csv('/content/bat.csv')
cric
```

```python
cric.columns
```

```python
df=pd.DataFrame({'Player':cric['Player'],'Team':cric['Team'],'Type':cric['Type'],'M':cric['M'],'Inn':cric['Inn'],})
df['Avg']=cric['Avg']
df
```

```python
df.groupby('Type').mean()
```

## H0: ODI  averge is 23 runs .
## H1: ODI average is more than  23 runs.

```python
sample=df.loc[df['Type']=='ODI']
samplec=sample.sample(165)
samplec
```

```python
samplec.Avg.mean()
```

z test

```python
#standard deviation of the AVG
sd=np.std(samplec['Avg'])
sd
```

```python
samplec.std()
```

```python
xbar=samplec.Avg.mean()
mu=337.09
n=165 #sample size
alpha=0.05
```

```python
xbar
```

```python
zvalue=(xbar-mu)/(sd/np.sqrt(n))
zvalue
```

```python
z_critical = st.norm.ppf(alpha)  #One tail Test
z_critical
```

```python
if (zvalue<z_critical and abs(zvalue)> abs(z_critical)):
  print("Reject Null Hypotheisis")
else:
  print("Fail to reject Null Hypothesis")
```

## using p value

```python
p_val = (1 - st.norm.cdf(abs(zvalue)))
p_val
```

```python
#comparing with alpha value
if p_val<alpha:
  print("Reject Null Hypotheisis")
else:
  print("Fail to reject Null Hypothesis")
```

## ODI average is more than 23 runs, i.e, we reject the Null Hypothesis

**T Test**

```python
tsamplec=sample.sample(25)
tsamplec
```

```python
t_sd=np.std(tsamplec['Avg'])
t_sd
```

```python
xbar=tsamplec.Avg.mean()
print(xbar)
mu=337.09
n=25 #sample size
alpha=0.05
```

```python
n = 25
degrees_of_freedom = n-1
t = (xbar - mu)/(t_sd / np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha, degrees_of_freedom)
t_critical
```

```python
if (t > t_critical):
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
```

## Both Z test and T test suggest the same result, The Null Hypothesis can be rejected.
